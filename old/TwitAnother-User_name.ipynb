```python
import pandas as pd
from io import BytesIO
tweetsdf = pd.DataFrame(columns=['created_at','id','full_text','urls','in_reply_to_status_id','in_reply_to_user_id','geo','coordinates','place','is_quote-status','retweet_count','favorite_count','reply_count','quote_count','conversation_id','possibly_sensitive'])
tweetsdf
```

```python
new_string = data.decode("utf-8")

#total tweets
startIndex = new_string.find('{\"tweets')
new_string = new_string[startIndex+13:]
endIndex = new_string.find('\":')
tweets = new_string[:endIndex]
#print(tweets)


while True:
    #created_at
    startIndex = new_string.find('{\"created_at')
    if(startIndex==-1):
        break
    new_string = new_string[startIndex+15:]
    endIndex = new_string.find('\", \"')
    created_at = new_string[:endIndex]
    #print(created_at)

    #id
    startIndex = new_string.find('\"id')
    new_string = new_string[startIndex+6:]
    endIndex = new_string.find(', \"')
    id = new_string[:endIndex]
    #print(id)

    #full_text
    startIndex = new_string.find('\"full_text')
    new_string = new_string[startIndex+14:]
    endIndex = new_string.find('\", \"')
    full_text = new_string[:endIndex]
    #print(full_text)

    #urls
    startIndex = new_string.find('\"urls')
    new_string = new_string[startIndex+10:]
    endIndex = new_string.find(', \"')
    urls = new_string[:endIndex]
    #print(urls)

    #in_reply_to_status_id
    startIndex = new_string.find('\"in_reply_to_status_id')
    new_string = new_string[startIndex+25:]
    endIndex = new_string.find(', \"')
    in_reply_to_status_id = new_string[:endIndex]
    #print(in_reply_to_status_id)

    #in_reply_to_user_id
    startIndex = new_string.find('\"in_reply_to_user_id')
    new_string = new_string[startIndex+23:]
    endIndex = new_string.find(', \"')
    in_reply_to_user_id = new_string[:endIndex]
    #print(in_reply_to_user_id)

    #geo
    startIndex = new_string.find('\"geo')
    new_string = new_string[startIndex+7:]
    endIndex = new_string.find(', \"')
    geo = new_string[:endIndex]
    #print(geo)

    #coordinates
    startIndex = new_string.find('\"coordinates')
    new_string = new_string[startIndex+15:]
    endIndex = new_string.find(', \"')
    coordinates = new_string[:endIndex]
    #print(coordinates)

    #place
    startIndex = new_string.find('\"place')
    new_string = new_string[startIndex+9:]
    endIndex = new_string.find(', \"')
    place = new_string[:endIndex]
    #print(place)

    #is_quote_status
    startIndex = new_string.find('\"is_quote_status')
    new_string = new_string[startIndex+19:]
    endIndex = new_string.find(', \"')
    is_quote_status = new_string[:endIndex]
    #print(is_quote_status)

    #retweet_count
    startIndex = new_string.find('\"retweet_count')
    new_string = new_string[startIndex+17:]
    endIndex = new_string.find(', \"')
    retweet_count = new_string[:endIndex]
    #print(retweet_count)

    #favorite_count
    startIndex = new_string.find('\"favorite_count')
    new_string = new_string[startIndex+18:]
    endIndex = new_string.find(', \"')
    favorite_count = new_string[:endIndex]
    #print(favorite_count)

    #reply_count
    startIndex = new_string.find('\"reply_count')
    new_string = new_string[startIndex+15:]
    endIndex = new_string.find(', \"')
    reply_count = new_string[:endIndex]
    #print(reply_count)

    #quote_count
    startIndex = new_string.find('\"quote_count')
    new_string = new_string[startIndex+15:]
    endIndex = new_string.find(', \"')
    quote_count = new_string[:endIndex]
    #print(quote_count)

    #conversation_id
    startIndex = new_string.find('\"conversation_id')
    new_string = new_string[startIndex+19:]
    endIndex = new_string.find(', \"')
    conversation_id = new_string[:endIndex]
    #print(conversation_id)

    #possibly_sensitive
    startIndex = new_string.find('\"possibly_sensitive')
    new_string = new_string[startIndex+22:]
    endIndex = new_string.find(', \"')
    possibly_sensitive = new_string[:endIndex]
    #print(possibly_sensitive)

    new_row = {'created_at': created_at, 'id': id, 'full_text': full_text, 'urls': urls, 'in_reply_to_status_id': in_reply_to_status_id,
               'in_reply_to_user_id': in_reply_to_user_id, 'geo': geo, 'coordinates': coordinates, 'place': place,
               'is_quote_status': is_quote_status, 'retweet_count': retweet_count, 'favorite_count': favorite_count,
               'reply_count': reply_count, 'quote_count': quote_count, 'conversation_id': conversation_id,
               'possibly_sensitive': possibly_sensitive}


    #append row to the dataframe
    tweetsdf = tweetsdf.append(new_row, ignore_index=True)



print("Done.")
```
