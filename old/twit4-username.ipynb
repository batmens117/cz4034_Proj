```python
#tweets
conn = http.client.HTTPSConnection("twitter154.p.rapidapi.com")

headers = {
    'X-RapidAPI-Key': "9c48575bd4mshc075073bb9ffd62p1e7aa7jsn60e737d8bf25",
    'X-RapidAPI-Host': "twitter154.p.rapidapi.com"
    }
conn.request("GET", "/user/tweets?username=elonmusk&limit=100", headers=headers)

res = conn.getresponse()
tweets = res.read()

tweetsdf = pd.read_csv(BytesIO(tweets))
```

```python
print(tweets.decode("utf-8"))
```

```python
tweetsdf = pd.DataFrame(columns=['tweet_id','creation_date','text','media_url'])
```

```python
tweetsdf
```

```python
og_string = tweets.decode("utf-8")
#print(og_string.split("{\"tweet_id\":"))
```

```python
new_string = tweets.decode("utf-8")

while True:
    #tweet_id
    startIndex = new_string.find('{\"tweet_id')
    if(startIndex==-1):
        break
    new_string = new_string[startIndex+13:]
    endIndex = new_string.find('\",\"')
    tweet_id = new_string[:endIndex]
    #print(tweet_id)

    #creation_date
    startIndex = new_string.find('\"creation_date')
    new_string = new_string[startIndex+17:]
    endIndex = new_string.find('\",\"')
    creation_date = new_string[:endIndex]
    #print(creation_date)

    #text
    startIndex = new_string.find('\"text')
    new_string = new_string[startIndex+8:]
    endIndex = new_string.find('\",\"')
    text = new_string[:endIndex]
    #print(text)

    #media_url
    startIndex = new_string.find('\"media_url')
    new_string = new_string[startIndex+12:]
    endIndex = new_string.find(',\"')
    media_url = new_string[:endIndex]
    #print(media_url)

    new_row = {'tweet_id': tweet_id, 'creation_date': creation_date, 'text': text, 'media_url': media_url}
    #append row to the dataframe
    tweetsdf = tweetsdf.append(new_row, ignore_index=True)

print("done.")
```

```python
tweetsdf.to_csv('elon_tweets.csv')
```
